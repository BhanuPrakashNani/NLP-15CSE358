```
from nltk.tokenize import word_tokenize
```

```
file_1 = open('chatbot.txt',mode='r')
tokens = []
for line in file_1:
  tokens.append(word_tokenize(line))
```

```
tokens
```

```
#Q4 and Q5
from nltk.corpus import stopwords
stop_words = list(stopwords.words('english'))
filtered = []
cnt1 = cnt2 = 0
for token in tokens:
  for tok in token:
    if tok not in stop_words:
      cnt1 += 1
      filtered.append(tok)
    else:
      cnt2 += 1
print(filtered)
print("Stop Words: ", cnt2)
print("Non Stop Words: ", cnt1)
```

```
#Q6
from random import randint
from translate import Translator
translator = Translator(to_lang="Hindi")
for i in range(10):
  t1 = randint(0, len(tokens)-1)
  t = randint(0, len(tokens[t1])-1)
  original = tokens[t1][t]
  trans = translator.translate(original)
  print("Original word: "+original+" , Translated: "+trans )
```

```
#Q7
from spellchecker import SpellChecker
spell = SpellChecker()
#spell.word_frequency.load_words()
for token in tokens:
  mispelled = spell.unknown(token)
  for word in mispelled:
    print(word, "Correction: ", spell.correction(word))
    print("Candidates: ", spell.candidates(word))
    if word in token:
      token.remove(word)
print(tokens)
```
